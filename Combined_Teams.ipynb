```python
# ODI
# create a list of year values for which you want to scrape data
year_list = [2010,2011,2012,2013,2014,2015,2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

# create an empty DataFrame to store the combined data
ODI_df = pd.DataFrame(columns=['Year', 'Team 1', 'Team 2', 'Winner 3', 'Margin', 'Ground', 'Match_Date', 'ScoreCard Text', 'ScoreCard URL'])

# loop over the year values
for year in year_list:
    # generate the URL for the current year
    url = f'https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?class=2;id={year};type=year'

    # send a request to the URL and get the response
    response = requests.get(url)

    # create a BeautifulSoup object
    soup = BeautifulSoup(response.text, 'html.parser')

    # find the table you want to extract data from
    table = soup.find('table')

    # extract the table data
    table_data = []
    for row in table.find('tbody').find_all('tr'):
        row_data = []
        for i, cell in enumerate(row.find_all('td')):
            # check if this is the 7th column
            if i == 6:
                # get the hyperlink text and URL
                hyperlink = cell.find('a')
                if hyperlink:
                    row_data.append(hyperlink.text)
                    row_data.append(hyperlink['href'])
                else:
                    row_data.append('')
                    row_data.append('')
            else:
                row_data.append(cell.text)
        row_data.insert(0, year)  # insert the year value as the first column
        table_data.append(row_data)

    # create a pandas DataFrame to store the data for the current year
    df = pd.DataFrame(table_data, columns=['Year', 'Team 1', 'Team 2', 'Winner 3', 'Margin', 'Ground', 'Match_Date', 'ScoreCard Text', 'ScoreCard URL'])

    # append the data for the current year to the combined DataFrame
    ODI_df = pd.concat([ODI_df, df], ignore_index=True)

# export the combined DataFrame to a CSV file
#combined_df.to_csv('all_years.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# define the years you want to extract data for
years = [2023, 2022, 2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014]

# create an empty list to store the dataframes for each year
year_dfs = []

# loop through each year and extract the data
for year in years:
    url = f'https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?class=2;id={year};type=year'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table')

    # extract the table data
    table_data = []
    for row in table.find_all('tr'):
        row_data = []
        for cell in row.find_all('td'):
            if cell.find('a') and 'Scorecard' in cell.find('a').text:
                row_data.append(cell.find('a').get('href'))
            else:
                row_data.append(cell.text)
        table_data.append(row_data)

    # create a pandas DataFrame to store the data
    df = pd.DataFrame(table_data[1:], columns=['Team 1', 'Team 2', 'Winner', 'Margin', 'Ground', 'Match_Date', 'Scorecard'])

    # add a 'Year' column to the dataframe
    df['Year'] = year

    # add the dataframe to the list
    year_dfs.append(df)

# concatenate the dataframes for each year into a single dataframe
combined_df4 = pd.concat(year_dfs, ignore_index=True)

# export the combined dataframe to a CSV file
#combined_df4.to_csv('combined_years.csv', index=False)
```
